```
# Databricks notebook source
from pyspark.sql import SparkSession
from pyspark.sql.functions import col, lit, udf,avg,min,max
from pyspark.sql.types import IntegerType,DoubleType
spark = SparkSession.builder.appName("groupBy - Assignment").getOrCreate()

# COMMAND ----------
df = spark.read.options(header='True', inferSchema='True').csv('/FileStore/tables/StudentData.csv')
df.show(5)
```

```
# Total # of students enrolled in each course
df.select(df.name,df.course).groupBy(df.course).count().show()
```

```
# Total # of male and female enrolled in each course
df.select(df.gender,df.course).groupBy(df.course,df.gender).count().sort(df.course,df.gender).show()
```

```
# Total Marks achieved by each gender in each course

df.select(df.gender,df.course,df.marks).groupBy(df.course,df.gender).sum().sort(df.course,df.gender.desc()).show()
```

```
# Display min,max and avg marks obtained by each age group.
df.select(df.age,df.marks).groupBy(df.age).agg(min('marks').alias('min_marks'),max('marks').alias('max_marks'),avg('marks').alias('avg_marks')).show()
```
