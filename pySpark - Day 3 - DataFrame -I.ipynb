```
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName("SparkDataFrame").getOrCreate() # Only one spark session is allowed.
```

```
#df = spark.read. option("inferSchema",True).option("header",True).csv('/FileStore/tables/StudentData.csv')
df = spark.read. options(inferSchema=True,header=True, delimiter = ',').csv('/FileStore/tables/StudentData.csv') #This is more efficient way for the options
df.show()
```
